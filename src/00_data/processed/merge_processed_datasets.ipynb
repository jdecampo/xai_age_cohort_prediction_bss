```python
# Import libraries
import warnings
import pandas as pd

warnings.filterwarnings('ignore')
```

```python
# File paths
weather_2019 = "path_flat_file_/weather_daily_2019_24.csv"
bike_trips_2019 = "path_flat_file_/bike_usage_2019.csv"
working_calendar_2019 = "path_flat_file_/Working_Calendar_FY2019.csv"
md_bike_stations = "path_flat_file_/md_bike_station_2019.csv"
full_dataset_path = "path_flat_file_/full_dataset_2019.csv"
```

```python
# Load preprocess datasets
bike_trips = pd.read_csv(bike_trips_2019, header = 0, sep = ';')
md_bike_station_df = pd.read_csv(md_bike_stations , header = 0, sep = ';')
weather_2019 = pd.read_csv(weather_2019, header = 0, sep = ';')
calendar = pd.read_csv(working_calendar_2019, header = 0, sep = ';')
```

## Changes to Bike trips dataset

```python
# Adding station weather id 24 Casa de Campo to bike_trips
bike_trips['id_weather'] = 24

bike_trips.dtypes
```

## Changes weather dataset

```python
# Remove columns
weather_2019 = weather_2019.drop(columns=['PROVINCIA', 'MUNICIPIO'])

# Rename columns
weather_2019.rename(columns={'ESTACION': 'id_weather','ANO': 'year', 'MES': 'month', 'DAY': 'day'}, inplace=True)
display(weather_2019.head())
```

## Merging bike_trips + weather dataset

```python
bike_trips_weather = pd.merge(bike_trips, weather_2019, on=["id_weather","year","month","day"])
print(bike_trips.shape)
print(bike_trips_weather.shape)
```

## Merging bike_trips_weather + calendar dataset

```python
pd.set_option('display.max_columns', None)
print(bike_trips_weather.shape)
full_dataset = pd.merge(bike_trips_weather, calendar, on=["year","month","day"])
print(full_dataset.shape)
display(full_dataset.head())
```

```python
full_dataset.head()
```

## Merging Full dataset + Master Data Bike Station

```python
# Selecting relevant subset 'unplug_station_tot_bases' 'idunplug_latitude' and 'idunplug_longitude' columns from md_bike_stations dataframe
md_bike_unplug_stations_subset = md_bike_station_df[['idunplug_station', 'unplug_station_tot_bases', 'idunplug_latitude', 'idunplug_longitude']]

# Merge the dataframes based on 'idunplug_station' column
merged_df = pd.merge(full_dataset, md_bike_unplug_stations_subset, on='idunplug_station', how='left')

display(merged_df.head())
print(f"Merged DF shape: {merged_df.shape}")


# Selecting relevant subset 'plug_station_tot_bases' 'idplug_latitude' and 'idplug_longitude' columns from md_bike_stations dataframe
md_bike_plug_stations_subset = md_bike_station_df[['idplug_station', 'plug_station_tot_bases', 'idplug_latitude', 'idplug_longitude']]

# Merge the dataframes based on 'idplug_station' column
full_dataset_df = pd.merge(merged_df, md_bike_plug_stations_subset, on='idplug_station', how='left')

print(f"Final Full Dataset shape: {full_dataset_df.shape}")
display(full_dataset_df.head())
```

```python
# Export compiled datasets
full_dataset_df.to_csv('full_dataset_2019.csv', index=False, sep = ';') 
```
